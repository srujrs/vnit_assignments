```python
import sklearn
from numpy import where
from sklearn.datasets import make_classification
from matplotlib import pyplot
from numpy import unique
```

<h1>Synthetic Clustering Dataset</h1><hr>

```python
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, 
                           n_clusters_per_class=1, random_state=4)
print("Shape of X is ",X.shape)
print("Shape of y is ",y.shape)
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
```

<h1>Initial Scatter Plot</h1><hr>

```python
for class_value in range(3):
    row_ix = where(y == class_value)
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```

<h1>AGNES Model</h1><hr>

```python
model_agnes = AgglomerativeClustering(n_clusters=3)
yhat = model_agnes.fit_predict(X)
clusters = unique(yhat)
for cluster in clusters:
    row_ix = where(yhat == cluster)
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```

<h1>DBSCAN Model</h1><hr>

```python
model_dbc = DBSCAN(eps=0.20, min_samples=11)
yhat = model_dbc.fit_predict(X)
clusters = unique(yhat)
for cluster in clusters:
    row_ix = where(yhat == cluster)
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```

<h1>K-Means Model</h1><hr>

```python
model_kmeans = KMeans(n_clusters = 3)
model_kmeans.fit(X)
yhat = model_kmeans.predict(X)
clusters = unique(yhat)
for cluster in clusters:
    row_ix = where(yhat == cluster)
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
pyplot.show()
```
