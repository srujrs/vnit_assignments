```python
import matplotlib.pyplot as plt
import numpy as np

divisions = ["Male","Female"]
math_avg_marks = [df[df["gender"] == "male"]["math score"].mean(), 
                  df[df["gender"] == "female"]["math score"].mean()]

reading_avg_marks = [df[df["gender"] == "male"]["reading score"].mean(), 
                     df[df["gender"] == "female"]["reading score"].mean()]

writing_avg_marks = [df[df["gender"] == "male"]["writing score"].mean(), 
                     df[df["gender"] == "female"]["writing score"].mean()]

index = np.arange(len(divisions))
width = 0.2

plt.bar(index, math_avg_marks, width, color = "blue", label = "Math score AVG")
plt.bar(index + width, reading_avg_marks, width, color = "green", label = "Reading score AVG")
plt.bar(index + 2*width, writing_avg_marks, width, color = "red", label = "Writing score AVG")

plt.title("Male vs Female AVG scores")

plt.ylabel("Marks")
plt.xlabel("Gender")
plt.xticks(index + width, divisions)
plt.yticks(np.arange(0,110,10))

plt.legend(loc = "best")
plt.show()
```

```python
df1 = df.copy(deep=True)
df1.drop(columns = ["lunch","reading score","writing score"], inplace = True)
df1.head()
```

```python
df1.replace(to_replace=["group A", "group B", "group C", "group D", "group E"], value=[1,2,3,4,5], inplace = True)
df1.replace(to_replace=["none", "completed"], value=[1,2], inplace = True)
df1.replace(to_replace=["associate's degree", "bachelor's degree", "high school", "master's degree", "some college", "some high school"], value=[1,2,3,4,5,6], inplace = True)
df1.replace(to_replace=["male", "female"], value=[1,2], inplace = True)
df1.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

# input columns
x = df1.drop(columns=['math score'])
# output columns
y = df1['math score']

# splitting dataset into test and train
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.1)

# creating and training model
model = LinearRegression().fit(X_train,y_train)

predictions = model.predict(X_test)

print(type(y_test))
print(type(predictions))

# checking accuracy of scores
score = r2_score(y_test.astype(float),predictions)
error = mean_squared_error(y_test.astype(float),predictions)
print(score, error)
```
